```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from datetime import datetime
```

# Behest Plots

```python
# Define lists to store the data
timestamps = []
usr_cpu_usages = []
disk_writes = []
writesKB = []
memUsed = []

# CPU data
with open('./getBehest/ip-172-31-30-108-20230611.cpu', 'r') as file:
    for _ in range(15):
        next(file)
    for line in file:
        values = line.split()
        timestamps.append(values[1])
        usr_cpu_usages.append(int(values[2]))

# Disk data
with open('./getBehest/ip-172-31-30-108-20230611.dsk', 'r') as file:
    for _ in range(15):
        next(file)
    for line in file:
        values = line.split()
        disk_writes.append(int(values[7]))
        writesKB.append(int(values[9]))

# Memory data
with open('./getBehest/ip-172-31-30-108-20230611.numa', 'r') as file:
    for _ in range(15):
        next(file)
    for line in file:
        values = line.split()
        memUsed.append(int(values[2])/1000000)

first_timestamp = datetime.strptime(timestamps[0], "%H:%M:%S")
timestamps_seconds = [(datetime.strptime(timestamp, "%H:%M:%S") - first_timestamp).total_seconds() for timestamp in timestamps]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(17, 5))

axs[0].plot(timestamps_seconds, usr_cpu_usages)
axs[0].set_title('CPU Usage Over Time')
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('CPU Usage (%)')

axs[1].plot(timestamps_seconds, disk_writes)
axs[1].set_title('Disk Usage Over Time')
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Disk Usage (%)')

axs[2].plot(timestamps_seconds, memUsed)
axs[2].set_title('Memory Usage Over Time')
axs[2].set_xlabel('Time (s)')
axs[2].set_ylabel('Memory Usage (GB)')

plt.tight_layout()
plt.show()
```
